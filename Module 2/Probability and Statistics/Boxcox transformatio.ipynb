#### Box-cox transformation
Reference : https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html

```python
from scipy import stats
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5   #  creating a random variable from log gamma distribution (This is one pareto)
prob = stats.probplot(x, dist=stats.norm, plot=ax1)# Q-Q plot between X- log gamma, y -normal
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
```

```python
fig = plt.figure()
ax2 = fig.add_subplot(212)
xt, _ = stats.boxcox(x)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
```

### Take aways:
Here most of point lies close to straight line so  we can conclude xt is Guassian distributed.

xt is received from the boxcox of x which is log gamma distribution, a type of pareto distribution.

xt, _ = stats.boxcox(x) calculated the Î» internally
